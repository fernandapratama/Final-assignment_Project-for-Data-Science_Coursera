```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import urllib.request
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
```

```python
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html5lib")
type(soup)
```

```python
# display structure html sequence
for tag in soup.recursiveChildGenerator():
    if tag.name:
        print(tag.name)
```

### WebScrapping TABLE

```python
temp = soup.find_all('table')[1]

table2 = temp.find_all("tr")
for row in table2[1:-1]:
    print(row.find_all("td"))
```

```python
#Create Empty DataFrame
tesla_revenue = pd.DataFrame(columns=["Date", "Revenue"])

for row in table2[1:-1]:
    tesla_revenue = tesla_revenue.append({"Date":row.find_all("td")[0].text, "Revenue":row.find_all("td")[1].text}, 
                                         ignore_index=True)
    #print(row.find_all("td")[1].text)
    
tesla_revenue
```
